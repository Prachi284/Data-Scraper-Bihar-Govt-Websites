```python
#Extracting the block wise data
import pandas as pd
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import Select

# Set the path to your chromedriver.exe
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
wait = WebDriverWait(driver, 10)

# Navigate to the homepage
homepage_url = "http://52.172.141.50/slwm/Dashboard.aspx"
driver.get(homepage_url)

cmb_fy_dropdown = driver.find_element(By.ID, 'cmb_fy')
select = Select(cmb_fy_dropdown)

# Select the desired value
select.select_by_visible_text("2022-2023")
# Find and click on the link
link_element = driver.find_element(By.ID, 'lnk_tot_swachhta_karmi')
link_element.click()

for dis in districts:
    district_link = driver.find_element(By.LINK_TEXT, dis)
    district_link.click()
    # Create an Excel writer object for each district
    with pd.ExcelWriter(f'{dis}_karmiii_data2223.xlsx', engine='xlsxwriter') as writer:
        for block in block_name:
            try:
                block_link = driver.find_element(By.LINK_TEXT, block)
                # Use explicit wait to ensure the element is present before interacting with it
                wait.until(EC.element_to_be_clickable((By.LINK_TEXT, block)))
                block_link.click()

                # Use a more targeted XPath expression to get the outerHTML of a specific element
                elem = driver.find_element(By.XPATH, "//body")
                source_code = elem.get_attribute("outerHTML")
                soup = BeautifulSoup(source_code, 'lxml')
                table = soup.find('table')
                table_rows = table.find_all('tr')
                world_titles = table.find_all('th')
                world_table_titles = [title.text for title in world_titles]

                data_rows = []
                for row in table_rows:
                    row_data = [data.text.strip() for data in row.find_all('td')]
                    data_rows.append(row_data)

                # Create a DataFrame from the list of rows
                df = pd.DataFrame(data_rows, columns=world_table_titles)

                # Save the DataFrame to a new sheet in the Excel file
                sheet_name = f'{block}'
                df.to_excel(writer, sheet_name=sheet_name, index=False)

                # Navigate back to the district level
                driver.execute_script("window.history.go(-1)")

            except Exception as e:
                print(f"Error processing {dis} - {block}: {str(e)}")

        # Navigate back to the main page for the next district
        driver.execute_script("window.history.go(-1)")
```

```python
#we are working on this
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import Select

# Set the path to your chromedriver.exe
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Navigate to the homepage
homepage_url = "http://52.172.141.50/slwm/Dashboard.aspx"
driver.get(homepage_url)
WebDriverWait(driver, 250000)


# Select Financial Year
cmb_fy_dropdown = driver.find_element(By.ID, 'cmb_fy')
select = Select(cmb_fy_dropdown)
select.select_by_visible_text("2022-2023")

# Click on the total swachhta karmi link
link_element = driver.find_element(By.ID, 'lnk_tot_swachhta_karmi')
link_element.click()

# Loop through districts
#districts=['DARBHANGA']
# Add more districts as needed
for district in districts:
    try:
        district_link = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.LINK_TEXT, district)))
        district_link.click()

        # Loop through blocks
        #blocks=['KUSHESWAR ASTHAN EAST']
  # Add more blocks as needed
        for block in block_name:
            try:
                block_link = driver.find_element(By.LINK_TEXT, block)
                wait = WebDriverWait(driver, 200)
                block_link.click()

                with pd.ExcelWriter(f'{block}111_2223.xlsx', engine='xlsxwriter') as writer_block:
                    # Loop through panchayats
                  # Add more panchayats as needed
                    for panchayat in panchayat_list:
                        try:
                            panchayat_link = driver.find_element(By.LINK_TEXT, panchayat)
                            # Use explicit wait to ensure the element is present before interacting with it
                            wait = WebDriverWait(driver, 200)
                            wait.until(EC.element_to_be_clickable((By.LINK_TEXT, panchayat)))
                            panchayat_link.click()

                            elem = driver.find_element(By.XPATH, "//body")
                            source_code = elem.get_attribute("outerHTML")
                            soup = BeautifulSoup(source_code, 'lxml')
                            tables = soup.find_all('table')

                            data_frames = []
                            for idx, table in enumerate(tables):
                                try:
                                    table_rows = table.find_all('tr')
                                    world_table_titles = [th.text.strip() for th in table.find_all('th')]
                                    data_rows = [
                                        [data.text.strip() for data in row.find_all('td')] for row in table_rows]

                                    # Create a DataFrame from the list of rows
                                    df = pd.DataFrame(data_rows, columns=world_table_titles)

                                    # Save the DataFrame to the list
                                    data_frames.append(df)

                                except Exception as e_table:
                                    print(f"Error creating DataFrame for {district} - {block} - {panchayat}: {str(e_table)}")
                                    #driver.back()
                                    

                            # Save all DataFrames to a single Excel sheet
                            sheet_name = f'1_data'
                            for idx, df in enumerate(data_frames):
                                df.to_excel(writer_block, sheet_name=sheet_name, startrow=idx * (len(data_frames[0]) + 2), index=False)

                            # Navigate back to the block level
                            driver.execute_script("window.history.go(-1)")

                        except Exception as e_panchayat:
                            print(f"Error processing {district} - {block} - {panchayat}: {str(e_panchayat)}")

                # Navigate back to the district level
                driver.execute_script("window.history.go(-1)")

            except Exception as e_block:
                print(f"Error processing {district} - {block}: {str(e_block)}")

        # Navigate back to the main page for the next district
        driver.get(homepage_url)

    except Exception as e_district:
        print(f"Error processing {district}: {str(e_district)}")

# Close the browser window
driver.quit()
```
